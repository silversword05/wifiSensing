```python
from scipy.io import loadmat
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import plotly.graph_objects as go
from mpl_toolkits.mplot3d import Axes3D
import datetime, random
import scipy.interpolate as interp
import collections
```

```python



def run(i, sub = 20, plotError = False):
    print("experiment " + str(i))
    rates_csi = loadmat('data/data_'+str(i)+'/br_estimates_plomb_all.mat')
    times = loadmat('data/data_'+ str(i)+ '/timestamps_plomb.mat')
    times = pd.DataFrame(times["timestamps"], columns = ["st", "end"])
    br_rates = pd.DataFrame(rates_csi["br_estimates_all"])
    br_rates["st"] = times.st/1000
    br_rates["end"] = times.end/1000
    br_rates["avg"] = [(i+j)/2.0 for i,j in zip(br_rates["end"], br_rates["st"])]
    br_rates["start"] = br_rates["st"].apply(datetime.datetime.fromtimestamp)
    br_rates["ended"] = br_rates["end"].apply(datetime.datetime.fromtimestamp)


    vernier = loadmat('data/data_'+str(i)+'/breathing_ground.mat')
    vernier_times = loadmat('data/data_'+str(i)+'/timestamps_ground.mat')

    breathing_rates = []
    for j in range(len(vernier["breathing_ground"])):
        breathing_rates.append(vernier["breathing_ground"][j][0])

    end = []
    start = []
    for time in vernier_times["timestamps"]:
        start.append(time[0])
        end.append(time[1])
    ground_truth = pd.DataFrame({"rate": breathing_rates, "start":start, "end":end})
    ground_truth["st"] = ground_truth["start"].apply(datetime.datetime.fromtimestamp)
    ground_truth["en"] = ground_truth["end"].apply(datetime.datetime.fromtimestamp)
    ground_truth["avg"] = [(i+j)/2.0 for i,j in zip(ground_truth["end"], ground_truth["start"])]
    rate_coefs = interp.splrep(ground_truth.avg, ground_truth.rate)
    smooth = interp.splev(ground_truth.avg, rate_coefs, der=0, ext=0)
    err = sum(abs(smooth - ground_truth.rate))/len(ground_truth.rate)
    print("error of smooth: " + str(err))
    truth = interp.splev(br_rates.avg, rate_coefs, der=0, ext=0)
    errors = []
    for col in range(208):
        errors.append((sum(abs(br_rates[col] - truth))/len(br_rates[col]), col))

    errors.sort()
    if plotError:
        print(errors[:5])
        for error in errors[:5]:    
            plt.plot(br_rates.avg, br_rates[error[1]], label = "estimate")
            plt.plot(br_rates.avg, truth, label = "ground truth")
            plt.legend()
            plt.title("br estimates sub " + str(error[1]))
            plt.show()
    else:    
        plt.plot(br_rates.avg, br_rates[sub], label = "estimate")
        plt.plot(br_rates.avg, truth, label = "ground truth")
        plt.legend()
        plt.title("br estimates sub " + str(sub))
        plt.show()
    return errors
```

```python

errors = []
for i in range(16):
    try:
        errors.append(run(i, sub = 20, plotError = False))
    except Exception as e: 
        print(e)
        print(i)


```

```python
subcarriers = {i:[] for i in range(208)}

for error in errors:
    for j,e in enumerate(error):
        subcarriers[e[1]].append(j)

subScore = [(k,np.mean(val)) for k,val in subcarriers.items()]
```

```python
subScore.sort(key = lambda x: x[1])
```

```python

for score in subScore[:5]:
    for i in range(16):
        try:
            run(i, sub = score[0], plotError = False)
        except Exception as e: 
            print(e)
            print(i)
```

```python
def importer(i):
    breath_matrix = loadmat('data/data_' + str(i) +'/ground_truth_12_5_2021_5min_' + str(i)+'_100ms.mat')
    #print(breath_matrix["RR_bpm"])
    breath_matrix = pd.DataFrame({"RR_bpm" : breath_matrix["RR_bpm"][0], "curr_time" :breath_matrix["curr_time"][0]})

    rates_csi = loadmat('data/data_'+str(i)+'/br_estimates_plomb.mat')
    times = loadmat('data/data_'+ str(i)+ '/timestamps_plomb.mat')
    rates_csi = rates_csi["br_estimates"]
    times = pd.DataFrame(times["timestamps"], columns = ["st", "end"])
    br_rates = pd.DataFrame(rates_csi, columns = ["br_rate"])
    br_rates["st"] = times.st/1000
    br_rates["end"] = times.end/1000
    br_rates["start"] = br_rates["st"].apply(datetime.datetime.fromtimestamp)
    br_rates["ended"] = br_rates["end"].apply(datetime.datetime.fromtimestamp)
    clean_breath = breath_matrix.dropna()
    clean_breath.reset_index(drop = True)
    clean_breath["time"] = [None]*len(clean_breath)
    clean_breath["time"] = clean_breath["curr_time"].apply(datetime.datetime.fromtimestamp)
    return (clean_breath, br_rates)
```

```python
results = []
for i in range(1,14):
    try:
        results.append(importer(i))
    except:
        print("exception " + str(i))
```

```python
for result in results:
    clean_breath = result[0]
    br_rates = result[1]
    plt.plot(clean_breath.curr_time,clean_breath.RR_bpm)
    plt.plot(br_rates.end, br_rates.br_rate)
    plt.show()
```
